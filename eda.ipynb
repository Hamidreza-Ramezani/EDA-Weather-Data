```python
new_column_names = {
    'Date/Time': 'DateTime',
    'Temp_C': 'Temperature',
    'Dew Point Temp_C': 'DewPoint',
    'Rel Hum_%': 'Humidity',
    'Wind Speed_km/h': 'WindSpeed',
    'Visibility_km': 'Visibility',
    'Press_kPa': 'Pressure',
    'Weather': 'Conditions'
}

df.rename(columns=new_column_names, inplace=True)
df.head()
```

## Identify the hottest and coolest days

```python
min_temp = df['Temperature'].min()
max_temp = df['Temperature'].max()
print(min_temp, max_temp)
```

## Visualize temperature trend over time using a line chart

```python
%matplotlib inline

fig, axes = plt.subplots(figsize=(12, 6))
```

```python
axes.plot(df['DateTime'], df['Temperature'], label='Temperature')
axes.set_xlabel('Date')
axes.set_ylabel('Temperature')
axes.set_title('Temperature over time')
axes.legend()
fig
```

## Scatter plot between Temperature and Humidity

```python
fig, axes = plt.subplots(figsize=(12,6))

# axes.plot(df['Temperature'], df['Humidity'], label='Temperature vs Humidity', type='scatter')

axes.scatter(df['Temperature'], df['Humidity'], label='Temperature vs Humidity', alpha=0.5)
axes.set_xlabel('Temperature')
axes.set_ylabel('Humidity')
axes.set_title('Temperature vs Humidity')
axes.legend()



# fig

```
